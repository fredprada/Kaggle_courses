### Underfitting and Overfitting

#### Experimenting With Different Models

*Scikit learn documentation:*<br>
http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

**Overfitting:** <br><br>
a decision tree with lots of levels (splits) will have several leaves. <br>
if there are 10 splits, there will be 2^10 (1024) leaves. <br>
this is bad because it will be unreliable for new data predictions.
<br>
<br><br>
**Underfitting:**<br><br>
when it fails to capture important distinctions and patterns in the data, it performns poorly in training data, and this phenomenon is called underfitting.






image.png

to find the *sweet spot*, we need to use *max_leaf_nodes* argument. <BR>
it is very sensitive and as we allow more leaves on the model, we get closer to the *overfitting* area on the graph

```python
# this is a utility function to help compare MAE scores from different values for max_leaf_nodes
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
max_leaf_nodes = [5, 50, 500, 5000]
max_leaf_nodes_list = []

for i in max_leaf_nodes:
  my_mae = get_mae(i, train_X, val_X, train_y, val_y)
  max_leaf_nodes_list.append(my_mae)
best_leaf_nodes = max_leaf_nodes_list.index(min(max_leaf_nodes_list))
print('the best leaf nodes is {}'.format(max_leaf_nodes[best_leaf_nodes]))
  # from these options, 500 is the best
```

**RECAP** <BR>
* **Overfitting**: 
  * capturing spurious patterns that won't recur in the future, leading to less accurate predictions, or
* **Underfitting**: 
  * failing to capture relevant patterns, again leading to less accurate predictions.

```python
# Fill in argument to make optimal size and uncomment
final_model = DecisionTreeRegressor(max_leaf_nodes=best_leaf_nodes, random_state=0)

# fit the final model and uncomment the next two lines
final_model.fit(X, y)
```

```python
val_predictions = final_model.predict(X)
mae = mean_absolute_error(y, val_predictions)
print(mae)
```

```python
predicted_prices = final_model.predict(X)
list_predict = []

for i in predicted_prices:
  list_predict.append(i)

print(list_predict[0:5])
print(list_y[0:5])
```
